# Draw the load and PV profiles

This notebook draws the plot for the visualization of the load and PV profiles.

```python
# packages and plot settings
from utils.load_data import load_case, load_measurement, load_load_pv
import matplotlib.pyplot as plt
from configs.config import sys_config

# Plot options
plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['font.size'] = 16
plt.rcParams['font.weight'] = 'bold'
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.constrained_layout.use'] = True
plt.rcParams['legend.loc'] = 'lower right'
plt.rcParams['legend.fontsize'] = 16

name = sys_config['case_name']
```

```python
case_class = load_case()
z_noise_summary, v_est_summary = load_measurement()
load_active, load_reactive, pv_active_, pv_reactive_ = load_load_pv()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
for i in range(load_active.shape[1]):
    ax[0].plot(load_active[:12*24,i])
    ax[1].plot(load_reactive[:12*24,i])
ax[0].title.set_text('Load Active')
ax[1].title.set_text('Load Reactive')
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
for i in range(pv_active_.shape[1]):
    ax[0].plot(pv_active_[:12*24,i])
    ax[1].plot(pv_reactive_[:12*24,i])
ax[0].title.set_text('PV Active')
ax[1].title.set_text('PV Reactive')
```
